#### Getting an updated dataframe of one off classes in the CS or Math department which now stay with only normal undergraduate classes

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.value_counts.html
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.index.html

```python
value_counts = CS_MATH_df["NBR"].value_counts() 
unique_values = value_counts[value_counts == 1].index 
one_class_df = CS_MATH_df[(CS_MATH_df['NBR'].isin(unique_values)) & (CS_MATH_df['NBR'] <=399)]

one_class_df 
```

#### I wanted to make a code using pandas that does the same thing as the google sheets does for this grade distribution online

https://www.w3schools.com/python/python_lambda.asp

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html

```python

# Define the mapping of letter grades to numeric values
grade_mapping = {
    'A+': 4.0,
    'A': 4.0,
    'A-': 3.7,
    'B+': 3.3,
    'B': 3.0,
    'B-': 2.7,
    'C+': 2.3,
    'C': 2.0,
    'C-': 1.7,
    'D': 1.0,
    'F': 0.0
}

grades = ["A+", "A", "A-", "B+", "B", "B-", "C+", "C", "C-", "D", "F"]

# Calculate the weighted average grade for each class
weighted_avg_grades = one_class_df[grades].apply(lambda row: sum(row[column] * grade_mapping[column] for column in one_class_df[grades].columns) / row.sum(), axis=1)

# Print the average letter grades for each class
print(weighted_avg_grades)
```
